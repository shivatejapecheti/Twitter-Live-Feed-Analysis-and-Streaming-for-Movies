### Function to Return Coordinates With Place Name:

```python
def get_coordinate(place):
    location = geolocator.geocode(place)
    return [location.latitude, location.longitude]
```

### Function To Add the Movie Name, Tweet_Type and Coordinates of location of Tweet

```python
tweet_count=0
for i in Temp_collection.find():
    T_count=0
    V_count=0
    try:
        # Movie Type:
        
        text = i.get('text', None)
        _id = i['id']
        V_count=text.count("varisu")+text.count("Varisu")+text.count("VARISU")
        T_count=text.count("thunivu")+text.count("Thunivu")+text.count("THUNIVU")
        if T_count>V_count:
            Temp_collection.update_one({'id': _id}, {'$set':{"movie":"Thunivu"}})
        elif V_count>T_count:
            Temp_collection.update_one({'id': _id}, {'$set':{"movie":"Varisu"}})
        elif V_count==T_count:
            Temp_collection.update_one({'id': _id}, {'$set':{"movie":"Thunivu & Varisu"}})
        else:
            Temp_collection.update_one({'id': _id}, {'$set':{"movie":None}})  
        
        # Tweet_Type analysis:
        
        blob = TextBlob(text)
        text_type=blob.sentiment.polarity
        if text_type>0:
            #positive
            Temp_collection.update_one({'id': _id}, {'$set':{"Tweet_Type":"Positive"}})
        elif text_type<0:
            #negative
            Temp_collection.update_one({'id': _id}, {'$set':{"Tweet_Type":"Negative"}})
        else:
            #neutral
            Temp_collection.update_one({'id': _id}, {'$set':{"Tweet_Type":"Neutral"}})
            
        # ADDING Coordinates to tweet with place:
        try:
            place= i.get('user', None)
            if place is not None:
                place = place.get('location', None)
            if place is not None:
                coordinate=get_coordinate(place)
                Temp_collection.update_one({'id':_id}, {'$set':{'coordinates':coordinate}})

        except:
            pass
        
    except:
        pass
    tweet_count+=1
    if tweet_count%5==0:
        print("Tweet_Count:",tweet_count)
```
